```python
# This script automates the entire process for liver segmentation.
# It handles dependency installation, data loading, model building,
# training, and evaluation in a single, runnable file.

import subprocess
import sys
import os
import zipfile
import glob
import re # Import the re module for regular expressions


def install_and_import(package):
    """
    Checks if a package is installed. If not, it installs the package
    and then imports it. This is useful for self-contained Colab scripts.
    """
    try:
        __import__(package)
        print(f"✅ '{package}' is already installed.")
    except ImportError:
        print(f"📦 '{package}' not found. Installing now...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        __import__(package)
        print(f"🎉 '{package}' installed successfully!")

# Check for and install required libraries
install_and_import('numpy')
install_and_import('tensorflow')
install_and_import('matplotlib')
install_and_import('sklearn')
install_and_import('google.colab')


import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive

# --- PROVIDED LOSS AND METRIC FUNCTIONS --- #
# Dice Loss: Better for imbalanced data than Binary Cross-Entropy
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice

# Dice Coefficient: The main evaluation metric (should be maximized, ideal value 1.0)
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return dice

# --- U-Net Model Definition ---
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model


# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary."


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    model.compile(optimizer='adam', loss=dice_loss, metrics=[dice_coef])

    # 6. Train the Model
    print("\n--- Training Model ---")
    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max')
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'dice_loss': dice_loss, 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model

        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```

```python
# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights
        weight_mask = tf.where(y_true == 1, class_weights[1], class_weights[0])
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

# Calculate class weights for BCE based on the dataset distribution
# This needs to be calculated *after* loading and concatenating the data
# For now, we will define the function and calculate the weights later in main()
# using the loaded 'masks' array.

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    return {0: weight_for_0, 1: weight_for_1}

# Modify the main function to calculate weights and use the new loss
# (This part will be done in the next step by modifying the main function)

# Placeholder for demonstrating the new loss function
# We will compile the model with the new loss in the main function
```

```python
# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights
        weight_mask = tf.where(y_true == 1, class_weights[1], class_weights[0])
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    return {0: weight_for_0, 1: weight_for_1}


# --- U-Net Model Definition ---
# (Keep the original build_unet_model function as it was not modified)
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary."

    # Calculate class weights after loading data
    class_weights = calculate_class_weights(masks)
    print(f"Calculated Class Weights: {class_weights}")


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    # Use the combined loss function with calculated class weights
    model.compile(optimizer='adam', loss=lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), metrics=[dice_coef])


    # 6. Train the Model
    print("\n--- Training Model ---")
    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max')
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            # Need to pass custom objects when loading a model with custom layers/functions
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'combined_loss': lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model


        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive
import os
import zipfile
import glob
import re

# --- PROVIDED LOSS AND METRIC FUNCTIONS --- #
# Dice Loss: Better for imbalanced data than Binary Cross-Entropy
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice

# Dice Coefficient: The main evaluation metric (should be maximized, ideal value 1.0)
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return dice

# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights
        weight_mask = tf.where(y_true == 1, class_weights[1], class_weights[0])
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    return {0: weight_for_0, 1: weight_for_1}


# --- U-Net Model Definition ---
# (Keep the original build_unet_model function as it was not modified)
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary."

    # Calculate class weights after loading data
    class_weights = calculate_class_weights(masks)
    print(f"Calculated Class Weights: {class_weights}")


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    # Use the combined loss function with calculated class weights
    model.compile(optimizer='adam', loss=lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), metrics=[dice_coef])


    # 6. Train the Model
    print("\n--- Training Model ---")
    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max')
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            # Need to pass custom objects when loading a model with custom layers/functions
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'combined_loss': lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model


        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive
import os
import zipfile
import glob
import re

# --- PROVIDED LOSS AND METRIC FUNCTIONS --- #
# Dice Loss: Better for imbalanced data than Binary Cross-Entropy
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice

# Dice Coefficient: The main evaluation metric (should be maximized, ideal value 1.0)
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return dice

# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights, ensuring they are float32
        weight_mask = tf.cast(tf.where(y_true == 1, class_weights[1], class_weights[0]), tf.float32)
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    # Return weights as float32
    return {0: np.float32(weight_for_0), 1: np.float32(weight_for_1)}


# --- U-Net Model Definition ---
# (Keep the original build_unet_model function as it was not modified)
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary."

    # Calculate class weights after loading data
    class_weights = calculate_class_weights(masks)
    print(f"Calculated Class Weights: {class_weights}")


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    # Use the combined loss function with calculated class weights
    model.compile(optimizer='adam', loss=lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), metrics=[dice_coef])


    # 6. Train the Model
    print("\n--- Training Model ---")
    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max')
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            # Need to pass custom objects when loading a model with custom layers/functions
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'combined_loss': lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model


        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive
import os
import zipfile
import glob
import re

# --- PROVIDED LOSS AND METRIC FUNCTIONS --- #
# Dice Loss: Better for imbalanced data than Binary Cross-Entropy
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice

# Dice Coefficient: The main evaluation metric (should be maximized, ideal value 1.0)
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return dice

# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights, ensuring they are float32
        weight_mask = tf.cast(tf.where(y_true == 1, class_weights[1], class_weights[0]), tf.float32)
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    # Return weights as float32
    return {0: np.float32(weight_for_0), 1: np.float32(weight_for_1)}


# --- U-Net Model Definition ---
# (Keep the original build_unet_model function as it was not modified)
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary." # Corrected syntax here

    # Calculate class weights after loading data
    class_weights = calculate_class_weights(masks)
    print(f"Calculated Class Weights: {class_weights}")


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    # Define the Adam optimizer with a custom learning rate
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

    # Use the combined loss function with calculated class weights
    model.compile(optimizer=optimizer, loss=lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), metrics=[dice_coef])


    # 6. Train the Model
    print("\n--- Training Model ---")
    # Define the learning rate scheduler callback
    lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_dice_coef',  # Monitor the validation Dice coefficient
        factor=0.1,               # Reduce learning rate by a factor of 10
        patience=5,               # Number of epochs with no improvement after which learning rate will be reduced
        mode='max',               # The monitoring quantity should be maximized
        verbose=1                 # Print a message when the learning rate is reduced
    )

    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max'),
        lr_scheduler # Add the learning rate scheduler to the callbacks
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            # Need to pass custom objects when loading a model with custom layers/functions
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'combined_loss': lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model


        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive
import os
import zipfile
import glob
import re

# --- PROVIDED LOSS AND METRIC FUNCTIONS --- #
# Dice Loss: Better for imbalanced data than Binary Cross-Entropy
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice

# Dice Coefficient: The main evaluation metric (should be maximized, ideal value 1.0)
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true = tf.cast(y_true, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    return dice

# Combined Weighted Binary Cross-Entropy and Dice Loss
def combined_loss(y_true, y_pred, alpha=0.5, beta=0.5, class_weights=None):
    """
    Combines Weighted Binary Cross-Entropy and Dice Loss.

    Args:
        y_true: Ground truth masks.
        y_pred: Predicted masks.
        alpha: Weight for Binary Cross-Entropy term.
        beta: Weight for Dice Loss term.
        class_weights: Dictionary or list of weights for each class in BCE.
                       If None, standard BCE is used.
    """
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Weighted Binary Cross-Entropy
    if class_weights is not None:
        bce = tf.keras.losses.BinaryCrossentropy(from_logits=False, reduction='none')(y_true, y_pred)
        # Apply class weights, ensuring they are float32 and have the same shape as bce
        weight_mask = tf.cast(tf.where(y_true == 1, class_weights[1], class_weights[0]), tf.float32)
        # Remove the last dimension from weight_mask to match bce shape
        weight_mask = tf.squeeze(weight_mask, axis=-1)
        weighted_bce = tf.reduce_mean(bce * weight_mask)
    else:
        weighted_bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)(y_true, y_pred)

    # Dice Loss
    smooth = 1e-6
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    dice = (2. * intersection + smooth) / (union + smooth)
    dice_loss_val = 1 - dice

    return alpha * weighted_bce + beta * dice_loss_val

def calculate_class_weights(masks):
    """Calculates class weights based on pixel distribution in masks."""
    total_pixels = masks.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels

    # Avoid division by zero if a class has no pixels
    weight_for_0 = (total_pixels / (2 * background_pixels)) if background_pixels > 0 else 1.0
    weight_for_1 = (total_pixels / (2 * liver_pixels)) if liver_pixels > 0 else 1.0

    # Return weights as float32
    return {0: np.float32(weight_for_0), 1: np.float32(weight_for_1)}


# --- U-Net Model Definition ---
# (Keep the original build_unet_model function as it was not modified)
def build_unet_model(input_shape):
    inputs = keras.Input(input_shape)
    # Contracting Path (Encoder)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = keras.layers.Dropout(0.1)(c1)
    c1 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = keras.layers.Dropout(0.1)(c2)
    c2 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = keras.layers.Dropout(0.2)(c3)
    c3 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = keras.layers.Dropout(0.2)(c4)
    c4 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)

    # Bridge
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = keras.layers.Dropout(0.3)(c5)
    c5 = keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # Expanding Path (Decoder)
    u6 = keras.layers.UpSampling2D((2, 2))(c5)
    u6 = keras.layers.concatenate([u6, c4])
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = keras.layers.Dropout(0.2)(c6)
    c6 = keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = keras.layers.UpSampling2D((2, 2))(c6)
    u7 = keras.layers.concatenate([u7, c3])
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = keras.layers.Dropout(0.2)(c7)
    c7 = keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = keras.layers.UpSampling2D((2, 2))(c7)
    u8 = keras.layers.concatenate([u8, c2])
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = keras.layers.Dropout(0.1)(c8)
    c8 = keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = keras.layers.UpSampling2D((2, 2))(c8)
    u9 = keras.layers.concatenate([u9, c1])
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = keras.layers.Dropout(0.1)(c9)
    c9 = keras.layers.Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# --- YOUR PATHS --- #
# IMPORTANT: Mount your Google Drive and update this path.
zip_file_path = "/content/drive/MyDrive/cnn_preprocessed.zip"
extracted_dir = "/content/cnn_preprocessed" # Directory to extract the zip file
data_dir_inside_zip = "cnn_npz" # The main folder inside the zip containing patient folders


def main():
    """Main function to run the entire pipeline."""

    # Mount Google Drive
    print("--- Mounting Google Drive ---")
    drive.mount('/content/drive')
    print("✅ Google Drive mounted successfully!")

    # Unzip the dataset
    print("\n--- Unzipping Dataset ---")
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_dir)
        print(f"✅ Dataset unzipped successfully to '{extracted_dir}'!")
    except FileNotFoundError:
        print(f"❌ Error: Zip file not found at '{zip_file_path}'. Please upload it and correct the path.")
        print("Make sure you have mounted Google Drive and the path is correct.")
        return
    except Exception as e:
        print(f"An error occurred while unzipping the file: {e}")
        return

    # 1. Load and Verify the Data
    print("\n--- Loading Data from Unzipped Folders ---")
    image_list = []
    mask_list = []
    data_root_path = os.path.join(extracted_dir, data_dir_inside_zip)

    if not os.path.exists(data_root_path):
        print(f"❌ Error: Data directory '{data_root_path}' not found after unzipping.")
        print("Please verify the structure of your zip file and the 'data_dir_inside_zip' variable.")
        return

    patient_folders = glob.glob(os.path.join(data_root_path, '*'))

    if not patient_folders:
        print(f"❌ Error: No patient folders found inside '{data_root_path}'.")
        print("Please check the structure of your unzipped data.")
        return

    print(f"Found {len(patient_folders)} patient folders.")

    # Load image and mask slices from each patient folder
    for patient_folder in patient_folders:
        if os.path.isdir(patient_folder):
            npz_files = sorted(glob.glob(os.path.join(patient_folder, '*.npz'))) # Sort files to maintain order
            if npz_files:
                patient_images = []
                patient_masks = []
                for npz_file in npz_files:
                    try:
                        data = np.load(npz_file)
                        # Use the keys 'image' and 'mask' as confirmed by the user
                        image_slice = data['image']
                        mask_slice = data['mask']

                        patient_images.append(image_slice)
                        patient_masks.append(mask_slice)
                    except KeyError as e:
                         print(f"❌ Error: Key {e} not found in {npz_file}. Please ensure the keys 'image' and 'mask' exist in all .npz files.")
                         # Skip loading data from this patient folder if keys are incorrect
                         patient_images = []
                         patient_masks = []
                         break # Exit the inner loop for this patient folder
                    except Exception as e:
                        print(f"❌ Error loading data from {npz_file}: {e}")
                        # Skip loading data from this patient folder if an error occurs
                        patient_images = []
                        patient_masks = []
                        break # Exit the inner loop for this patient folder


                if patient_images and patient_masks and len(patient_images) == len(patient_masks):
                    # Assuming slices need to be stacked to form a volume for the patient
                    # You might need to adjust axis depending on how slices are oriented
                    try:
                        image_volume = np.stack(patient_images, axis=0) # Adjust axis if needed
                        mask_volume = np.stack(patient_masks, axis=0) # Adjust axis if needed
                        image_list.append(image_volume)
                        mask_list.append(mask_volume)
                    except ValueError as e:
                         print(f"❌ Error stacking slices for patient {os.path.basename(patient_folder)}: {e}")
                         print("Ensure that all slices for a patient have compatible shapes for stacking.")


            else:
                print(f"⚠️ Warning: No .npz files found in {patient_folder}. Skipping.")


    if not image_list or not mask_list:
        print("❌ Error: No data loaded from patient folders. Please check file paths, structure, and the logic for separating images and masks.")
        print("Ensure that the keys 'image' and 'mask' are present in all .npz files and that slices for each patient have compatible shapes.")
        return

    # Concatenate data from all patients
    # This assumes all patient volumes have compatible shapes for concatenation after stacking slices.
    # You might need padding or resizing if shapes vary.
    try:
        images = np.concatenate(image_list, axis=0)
        masks = np.concatenate(mask_list, axis=0)
    except ValueError as e:
        print(f"❌ Error concatenating data from different patients: {e}")
        print("Ensure that image and mask volumes from different patients have compatible shapes after stacking slices.")
        return


    print("✅ Data loaded successfully!")
    print(f"Combined Images shape: {images.shape}")
    print(f"Combined Masks shape: {masks.shape}")

    # Verify shapes
    assert images.shape == masks.shape, "Shape mismatch between images and masks after concatenation."

    # Verify mask values
    unique_mask_values = np.unique(masks)
    print(f"Unique mask values: {unique_mask_values}")
    assert np.all(np.isin(unique_mask_values, [0, 1])), "Mask is not binary."

    # Calculate class weights after loading data
    class_weights = calculate_class_weights(masks)
    print(f"Calculated Class Weights: {class_weights}")


    # 2. Exploratory Data Analysis (EDA)
    print("\n--- Exploratory Data Analysis ---")
    # Plot sample images and masks
    # Assuming the data is now a 3D volume (num_samples, height, width) or (num_samples, depth, height, width)
    # Adjust plotting if data is 4D (e.g., includes a depth dimension per patient volume)
    num_samples_to_plot = min(5, images.shape[0]) # Plot up to 5 samples (slices) from the combined data
    plt.figure(figsize=(12, 6))
    for i in range(num_samples_to_plot):
        # Plot Image
        plt.subplot(2, num_samples_to_plot, i + 1)
        # Assuming images is (num_samples, height, width)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Sample Image {i+1}")
        plt.axis('off')

        # Plot Mask
        plt.subplot(2, num_samples_to_plot, num_samples_to_plot + i + 1)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f"Sample Mask {i+1}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Plot pixel value histogram
    plt.figure(figsize=(8, 5))
    plt.hist(images.flatten(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Histogram of Image Pixel Values")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.show()

    # Calculate class distribution
    total_pixels = images.size
    liver_pixels = np.sum(masks == 1)
    background_pixels = total_pixels - liver_pixels
    liver_percentage = (liver_pixels / total_pixels) * 100
    background_percentage = (background_pixels / total_pixels) * 100
    print(f"Class Distribution:")
    print(f"  - Liver pixels (1): {liver_pixels} ({liver_percentage:.2f}%)")
    print(f"  - Background pixels (0): {background_pixels} ({background_percentage:.2f}%)")

    # 3. Data Preparation for CNN
    print("\n--- Preparing Data for CNN ---")
    # Assuming the data is now a collection of 2D slices (num_samples, height, width)
    # If your model expects 3D volumes per patient, this splitting and reshaping needs adjustment.
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Add a channel dimension for the CNN (assuming 2D CNN slices)
    X_train = np.expand_dims(X_train, axis=-1)
    y_train = np.expand_dims(y_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    y_val = np.expand_dims(y_val, axis=-1)

    # Convert masks to float32
    y_train = y_train.astype(np.float32)
    y_val = y_val.astype(np.float32)

    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"X_val shape: {X_val.shape}")
    print(f"y_val shape: {y_val.shape}")

    # 4. Build a U-Net CNN Model
    print("\n--- Building U-Net Model ---")
    # The input shape should match the shape of a single image slice with a channel dimension
    if X_train.shape:
        input_shape = X_train.shape[1:] # Get shape excluding the batch size
        model = build_unet_model(input_shape)
        model.summary()
    else:
        print("❌ Error: Training data (X_train) is empty. Cannot build model.")
        return


    # 5. Compile the Model
    print("\n--- Compiling Model ---")
    # Define the Adam optimizer with a custom learning rate
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

    # Use the combined loss function with calculated class weights
    model.compile(optimizer=optimizer, loss=lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), metrics=[dice_coef])


    # 6. Train the Model
    print("\n--- Training Model ---")
    # Define the learning rate scheduler callback
    lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_dice_coef',  # Monitor the validation Dice coefficient
        factor=0.1,               # Reduce learning rate by a factor of 10
        patience=5,               # Number of epochs with no improvement after which learning rate will be reduced
        mode='max',               # The monitoring quantity should be maximized
        verbose=1                 # Print a message when the learning rate is reduced
    )

    callbacks = [
        keras.callbacks.EarlyStopping(patience=10, verbose=1),
        keras.callbacks.ModelCheckpoint('liver_unet.keras', verbose=1, save_best_only=True, monitor='val_dice_coef', mode='max'),
        lr_scheduler # Add the learning rate scheduler to the callbacks
    ]

    # Check if training data is available before fitting
    if X_train.shape[0] > 0:
        history = model.fit(X_train, y_train,
                            epochs=50,
                            batch_size=16,
                            callbacks=callbacks,
                            validation_data=(X_val, y_val))

        # 7. Evaluate the Model
        print("\n--- Evaluating Model ---")
        # Plot training history
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['dice_coef'], label='Training Dice Coef')
        plt.plot(history.history['val_dice_coef'], label='Validation Dice Coef')
        plt.title('Training and Validation Dice Coefficient')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Coef')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Load the best model
        try:
            # Need to pass custom objects when loading a model with custom layers/functions
            best_model = keras.models.load_model('liver_unet.keras', custom_objects={'combined_loss': lambda y_true, y_pred: combined_loss(y_true, y_pred, class_weights=class_weights), 'dice_coef': dice_coef})
            print("✅ Best model loaded from 'liver_unet.keras'.")
        except Exception as e:
            print(f"❌ Error loading the best model: {e}")
            print("Continuing with the last trained model.")
            best_model = model


        # Visualize predictions on validation data
        num_predictions_to_show = 3
        plt.figure(figsize=(15, 8))

        for i in range(num_predictions_to_show):
            sample_index = np.random.randint(0, len(X_val))
            test_image = X_val[sample_index:sample_index+1]
            predicted_mask = best_model.predict(test_image)[0]

            # Binarize the prediction
            predicted_mask_binary = (predicted_mask > 0.5).astype(np.float32)

            # Plot Original Image
            plt.subplot(3, num_predictions_to_show, i + 1)
            plt.imshow(test_image[0, :, :, 0], cmap='gray')
            plt.title(f"Original Image")
            plt.axis('off')

            # Plot True Mask
            plt.subplot(3, num_predictions_to_show, num_predictions_to_show + i + 1)
            plt.imshow(y_val[sample_index, :, :, 0], cmap='gray')
            plt.title(f"True Mask")
            plt.axis('off')

            # Plot Predicted Mask
            plt.subplot(3, num_predictions_to_show, 2 * num_predictions_to_show + i + 1)
            plt.imshow(predicted_mask_binary[:, :, 0], cmap='gray')
            plt.title(f"Predicted Mask")
            plt.axis('off')

        plt.tight_layout()
        plt.show()

        print("\n🎉 All tasks completed successfully! Check the plots above to see the results.")
    else:
        print("❌ Error: No training data available. Model training skipped.")

if __name__ == "__main__":
    main()
```
